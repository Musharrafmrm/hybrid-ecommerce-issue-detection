```python
# =============================================================================
# 07_Data_Cleaning_and_Preparation.ipynb
# Hybrid Machine Learning for E-Commerce Review Issue Detection
# =============================================================================

import pandas as pd
import re
import os
import zipfile

print("🔄 Checking and fixing data folder...\n")

# Create data folder
os.makedirs('data', exist_ok=True)

# Re-download Amazon dataset if file is missing
if not os.path.exists('data/Reviews.csv'):
    print("Downloading Amazon reviews again...")
    !kaggle datasets download -d snap/amazon-fine-food-reviews -f Reviews.csv --path data/ --unzip
    print("✅ Download completed!")
else:
    print("✅ Reviews.csv already exists.")

# Load the dataset
print("\nLoading dataset...")
df = pd.read_csv('data/Reviews.csv')

print(f"Original rows: {len(df):,}")

# Keep only useful columns
df = df[['Score', 'Summary', 'Text']].copy()

# Cleaning function
def clean_text(text):
    if not isinstance(text, str):
        return ""
    text = text.lower()
    text = re.sub(r'[^a-zA-Z\s]', '', text)      # Remove numbers and symbols
    text = re.sub(r'\s+', ' ', text).strip()     # Remove extra spaces
    return text

print("Cleaning review text...")
df['cleaned_text'] = df['Text'].apply(clean_text)

# Remove very short reviews
df = df[df['cleaned_text'].str.len() > 15]

print(f"After cleaning: {len(df):,} reviews")

# Save cleaned dataset
df.to_csv('data/cleaned_dataset.csv', index=False)
print("✅ Cleaned dataset saved as 'data/cleaned_dataset.csv'")

# Show summary
print("\nReview Score Distribution:")
print(df['Score'].value_counts().sort_index())

print("\nSample cleaned reviews:")
print(df[['Score', 'Summary', 'cleaned_text']].head(3))
```
